```python
from pyannote.audio import Pipeline

# loading
pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization")

# analyze audio
diarization = pipeline("audio.wav")

# print speaker diarization result
for turn, _, speaker in diarization.itertracks(yield_label=True):
    print(f"speaker {speaker}: {turn.start:.1f}s - {turn.end:.1f}s")
```

```python
import whisper

# load Whisper model
model = whisper.load_model("medium")

# voice to text
transcription = model.transcribe("audio.wav")

# print voice to text result
for segment in transcription["segments"]:
    print(f"{segment['start']:.1f}s - {segment['end']:.1f}s: {segment['text']}")
```

```python
# assume diarization  is speaker diarization result
# assume transcription is voice to text result

# convert speaker diarization result to time period list
diarization_segments = []
for turn, _, speaker in diarization.itertracks(yield_label=True):
    diarization_segments.append({
        "start": turn.start,
        "end": turn.end,
        "speaker": speaker
    })

# match voice to text result with speaker diarization result
for segment in transcription["segments"]:
    start = segment["start"]
    end = segment["end"]
    text = segment["text"]
    
    # find corresponding speaker
    for diarization_segment in diarization_segments:
        if diarization_segment["start"] <= start <= diarization_segment["end"] or \
   diarization_segment["start"] <= end <= diarization_segment["end"] or \
   (start <= diarization_segment["start"] and end >= diarization_segment["end"]):
            print(f"{diarization_segment['speaker']}: {text}")
            break
```
